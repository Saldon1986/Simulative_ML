#### **Задача 1**

Ваша задача написать класс `Extraction`, и определить в нем метод `from_csv`, который принимал бы путь к файлу и вытаскивал данные в виде списка объектов.

Для этого, здесь же, определите класс `Sale`.

**Решение**

Напишите свое решение ниже

```python
class Extraction:
  def from_csv(file_path):
    result = []
    with open(file_path, 'r') as file:
      headers = file.readline().strip().split()
      lines = [line.strip().split(',') for line in file.readlines()]
      for line in lines:
        sale = Sale(id=line[0], date=line[1], amount=float(line[2]), product=line[3])
        result.append(sale)
    return result

class Sale:
  def __init__(self, id, date, amount, product):
    self.__id = id
    self.__date = date
    self.__amount = float(amount)
    self.__product = product

  def get_id(self):
    return self.__id

  def get_date(self):
    return self.__date

  def get_amount(self):
    return self.__amount

  def get_product(self):
    return self.__product
```

```python
sales_list = Extraction.from_csv('sales_data.csv')
```

#### **Задача 2**

Напишите класс `Transformation`, который содержит два метода: `filter_by_date(sales_data, start_date, end_date)` и `filter_by_amount(sales_data, min_amount, max_amount)`.

**Решение**

Напишите свое решение ниже

```python
class Transformation:
  def filter_by_date(sales_data, start_date, end_date):
    return [l for l in sales_data if start_date <= l.get_date() <= end_date]

  def filter_by_amount(sales_data, min_amount, max_amount):
    return [l for l in sales_data if min_amount <= l.get_amount() <= max_amount]
```

#### **Задача 3**

Напишите класс `Analysis`, содержащий два метода: `calculate_total_sales` и `calculate_average_sales`. Возвращаемые значения округлите до двух знаков после запятой.

**Решение**

Напишите свое решение ниже

```python
class Analysis:
  def calculate_total_sales(sales_data):
    return sum([l.get_amount() for l in sales_data])

  def calculate_average_sales(sales_data):
    return sum([l.get_amount() for l in sales_data]) / len([l.get_amount() for l in sales_data])
```

#### **Задача 4**

Напишите класс `Loading` с методом `to_csv(sales_data, file_path)`, который запишет данные в `csv` файл.

**Решение**

Напишите свое решение ниже

```python
class Loading:
  def to_csv(sales_data, file_path, delimiter=','):
    with open(file_path, 'w', encoding='utf-8') as file:
      headers = ['id', 'date', 'amount', 'product']
      file.write(delimiter.join(headers) + '\n')
      for el in sales_data:
        row = [el.get_id(), el.get_date(), str(el.get_amount()), el.get_product()]
        file.write(delimiter.join(row) + '\n')
```
