### **Найдите самую эффективную платформу по средним значениям конверсий**

Cохраните результат в переменную `most_effective_platform`

````python
# Ваше решение
from collections import defaultdict

with open('campaign_data.csv', 'r') as f:
  headers = f.readline().rstrip().split(',')
  lines = [line.rstrip().split(',') for line in f.readlines()]

#создаем словарь Плафторма: список конверсий
# {'Платформа': [Конверсии]}
platform_conversations_dict = defaultdict(list)

for line in lines:
  platform = line[4]
  platform_conversations_dict[platform].append(int(line[9]))

platform_avg_dict = {
    platform: sum(conv_list) / len(conv_list)
    for platform, conv_list in platform_conversations_dict.items()
}

most_effective_platform = max(platform_avg_dict, key=platform_avg_dict.get)
````

### **Посчитайте долю затрат на рекламные кампании каждой платформы и каждого города**

Сохраните результат в файл `platform_city_results.txt`

Результат должен иметь такой же порядок групп, что и изначальный файл. А города внутри платформ должны быть отсортированы по убыванию доли затрат, а доля затрат должны быть округлена до 2 знаков после запятой.

Результат должен выглядеть примерно так:

```
Для группы Telegram:
- Город: Москва, доля затрат на рекламу: 16.46%
- Город: Калуга, доля затрат на рекламу: 16.2%
- Город: Казань, доля затрат на рекламу: 14.57%
- ...
Для группы Вконтакте:
- Город: Екатеринбург, доля затрат на рекламу: 17.43%
- Город: Новосибирск, доля затрат на рекламу: 14.86%
- Город: Санкт-Петербург, доля затрат на рекламу: 14.13%
...
```

````python
# Ваше решение
from collections import defaultdict

with open('campaign_data.csv', 'r') as f:
  headers = f.readline().rstrip().split(',')
  lines = [line.rstrip().split(',') for line in f.readlines()]

#создаем словарь Плафторма: словарь (Город: сумма затрат)
platform_dict = defaultdict(lambda: defaultdict(lambda: 0))

for line in lines:
  platform = line[4]
  city = line[6]
  platform_dict[platform][city] += int(line[5])

with open('platform_city_results.txt', 'w', encoding='utf-8') as res_f:
  for platform, cities in platform_dict.items():
    res_f.write(f'Для группы {platform}:\n')
    total_platform_budget = sum(cities.values())
    sorted_cities = dict(sorted(cities.items(), key=lambda x: x[1], reverse=True))
    platform_dict[platform] = sorted_cities
    for city, budget in sorted_cities.items():
      writeline = f'- Город: {city}, доля затрат на рекламу: {round(budget / total_platform_budget * 100, 2)}%\n'
      res_f.write(writeline)
  res_f.close()
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python
from collections import defaultdict
def calculate_average_budget(file_path, search_platfrom):
  with open(file_path, 'r') as f:
    headers = f.readline().rstrip().split(',')
    lines = [line.rstrip().split(',') for line in f.readlines()]

  #создаем словарь Плафторма: список конверсий
  # {'Платформа': [Конверсии]}
  platform_budget_dict = defaultdict(list)

  for line in lines:
    platform = line[4]
    platform_budget_dict[platform].append(int(line[5]))

  platform_avg_dict = {
      platform: round(sum(budget) / len(budget), 2)
      for platform, budget in platform_budget_dict.items()
  }
  return platform_avg_dict[search_platfrom]

    # Ваше решение


calculate_average_budget('campaign_data.csv', 'Google')
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python
from collections import defaultdict
from datetime import date, datetime, timedelta

def get_missing_campaign_dates(file_path):

    # Ваше решение
    with open('campaign_data.csv', 'r') as f:
      headers = f.readline().rstrip().split(',')
      lines = [line.rstrip().split(',') for line in f.readlines()]
    #dates with campaign_start
    campaign_dates = set()

    for line in lines:
      start_dt = date.fromisoformat(line[1])
      campaign_dates.add(start_dt)

    min_dt = min(campaign_dates)
    max_dt = max(campaign_dates)

    current = min_dt
    while current <= max_dt:
        if current not in campaign_dates:
            yield current
        current += timedelta(days=1)
````

````python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
from datetime import date

correct_answer = answers[5]

missing_dates = get_missing_campaign_dates('campaign_data.csv')
user_answer = []
for date in missing_dates:
    a = date.strftime('%Y-%m-%d')
    user_answer.append(a)

try:
    assert str(user_answer) == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
````

### **Напишите функцию, которая будет принимать на вход путь к файлу и возвращать сгруппированные данные по городам, содержащие суммарный бюджет и количество кликов**

Функция `group_campaign_data` должна возвращать список словарей, отсортированный по городам, в формате:

```python
[{'Город': 'Нижний Новгород',
  'Количество кликов': 616.0,
  'Суммарный бюджет': 27585},
 {'Город': 'Новосибирск',
  'Количество кликов': 733.0,
  'Суммарный бюджет': 33001},
  ...]
```

````python
from collections import defaultdict
def group_campaign_data(file_path):

  # Ваше решение
  with open('campaign_data.csv', 'r') as f:
    headers = f.readline().rstrip().split(',')
    lines = [line.rstrip().split(',') for line in f.readlines()]

  #создаем словарь (Город: budget)
  city_dict = defaultdict(lambda: [0.0, 0])

  for line in lines:
    clicks = float(line[8])
    city = line[6]
    budget = int(line[5])
    city_dict[city][0] += clicks
    city_dict[city][1] += budget

  res =[]

  for city, cl_bud in city_dict.items():
    res.append(
        {'Город': city,
         'Количество кликов': cl_bud[0],
         'Суммарный бюджет': cl_bud[1]}
    )

  res.sort(key=lambda x: x['Город'])

  return res
````
