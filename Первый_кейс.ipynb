**Примечание**

Не все файлы подходящие по названию, будут подходить по содержанию. Там может оказаться лишний столбец, например. Ориентируйтесь на столбцы из чека выше - это то, что вас интересует. Остальные столбцы можно просто отбросить.

**Важно**: разделителем файла на выходе должна быть запятая.

```python
def process_files(src_folder, dest_folder):
  import re
  import os, sys, glob

  src_folder = 'reports-main'
  dest_folder = 'comb_reports'
  #проверяем наличие папки с результатом
  os.makedirs(dest_folder, exist_ok=True)

  #отбираем в files только те, что проходят по смаске названий
  files = glob.glob(src_folder + '/*.csv')
  file_mask = r'reports-main/\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+.csv'
  files = [file for file in files if re.match(file_mask, file)]
  #сортируем от старых к новым по названию
  files.sort()

  #создаем заголовок итогового файла и пишем его в список полей для дальнейшей обработки - отсечения лишних данных в файлах
  #+ создаем итоговый файл с заголовком
  with open(files[0], 'r') as wfile:
    fields_res = wfile.readline().rstrip().split(';')
  with open(dest_folder + '/combined_data.csv', 'w') as dest_file:
    dest_file.write(','.join(fields_res) + '\n')

  #пробегаем по каждому файлу
  for file in files:
    with open(file, 'r') as wfile, open(dest_folder + '/combined_data.csv', 'a', newline='\n') as dest_file:
      #считываем строки файла
      lines = wfile.readlines()
      #определяем столбцы в файле
      fields = lines.pop(0).rstrip().split(';')
      #помещаем значения в список
      lines = [line.rstrip().split(';') for line in lines]
      #создаем словарь - название поля в файле: значение в файле
      full_lines = [dict(zip(fields, line)) for line in lines]
      #отбираем только те столбцы, которые есть в первом файле.
      #вернет KeyError если в первом файле полей больше, чем в одном из обрабатываемых. Подумать над значениями по умолчанию???
      filtered_lines = [','.join([full_line[k] for k in fields_res]) for full_line in full_lines]
      #пишем строки в итоговый файл
      for line in filtered_lines:
        dest_file.write(line+'\n')
```

```python
src_folder = 'reports-main'
dest_folder = 'comb_reports'
process_files(src_folder, dest_folder)
```
